# Shoreline Watch Data Visualizations

```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# Load Data
df = pd.read_excel('Shoreline Watch.xlsx')
```

```python
# Read first 5 Rows
df.head(5)
```

## Revenue by Year

Just like SQL, Pandas can GROUP BY columns by the same aggregate functions (SUM, MEAN, etc). That way, we can create Visualizations without even needing to touch SQL Code.

```python
# We GROUP BY Year and sum the column 'watch_price' to get Revenue by Year
# We then put the GROUP BY inside a Dataframe and reset_index because it defaults to year
gdf = pd.DataFrame(df.groupby(['year'])['watch_price'].sum()).reset_index()

# Set Figure Size
plt.figure(figsize = (12,8))

# Create Line Graph and use Custom Color
sns.pointplot(x = 'year', y = 'watch_price', data = gdf, color = '#4ab2f7')

# Graph and Axis Titles
plt.title('Revenue by Year', fontsize = 'xx-large', fontweight = 'bold')
plt.xlabel('Year', fontsize = 'x-large', fontweight = 'bold')
plt.ylabel('Revenue', fontsize = 'x-large', fontweight = 'bold')

# Set ticks to be Large for easier reading
plt.xticks(fontsize = 'large')
plt.yticks(fontsize = 'large')

plt.show()
```

# Quantity Sold by Year

```python
# Quantity Sold by Year
gdf = pd.DataFrame(df.groupby(['year'])['qty'].sum()).reset_index()

# Set Figure Size
plt.figure(figsize = (12,8))

# Create Bar Graph and use Color Palette
with sns.color_palette("husl", 2):
    sns.barplot(x = 'year', y = 'qty', data = gdf)

# Graph and Axis Titles
plt.title('Quantity Sold by Year', fontsize = 'xx-large', fontweight = 'bold')
plt.xlabel('Year', fontsize = 'x-large', fontweight = 'bold')
plt.ylabel('Quantity Sold', fontsize = 'x-large', fontweight = 'bold')

# Set ticks to be Large for easier reading
plt.xticks(fontsize = 'large')
plt.yticks(fontsize = 'large')

plt.show()
```

# Quantity Sold by Month

```python
# Quantity Sold by Year
gdf = pd.DataFrame(df.groupby(['MonthName'])['qty'].sum()).reset_index()

# Set Figure Size
plt.figure(figsize = (12,8))

# Create Bar Graph and Order Months Chronologically
sns.barplot(x = 'MonthName', y = 'qty', data = gdf,
            order=['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December'])

# Graph and Axis Titles
plt.title('Quantity Sold by Month', fontsize = 'xx-large', fontweight = 'bold')
plt.xlabel('Month', fontsize = 'x-large', fontweight = 'bold')
plt.ylabel('Quantity Sold', fontsize = 'x-large', fontweight = 'bold')


plt.yticks(fontsize = 'large')

plt.show()
```

# Total Quantity Sold by Brand (All Years)

```python
# Quantity Sold by Watch Brand
gdf = pd.DataFrame(df.groupby(['watch_brand'])['qty'].sum()).reset_index()

# Set Figure Size
plt.figure(figsize = (12,8))

# Create Bar Graph and use Custom Color
with sns.color_palette("husl", 2):
    sns.barplot(x = 'watch_brand', y = 'qty', data = gdf)

# Graph and Axis Titles
plt.title('Total Quantity Sold by Brand (All Years)', fontsize = 'xx-large', fontweight = 'bold')
plt.xlabel('Watch Brand', fontsize = 'x-large', fontweight = 'bold')
plt.ylabel('Quantity Sold', fontsize = 'x-large', fontweight = 'bold')

# Set ticks to be Large for easier reading
plt.xticks(fontsize = 'large')
plt.yticks(fontsize = 'large')

# Adjust Y-Axis
plt.ylim(0, 35000)

plt.show()
```
